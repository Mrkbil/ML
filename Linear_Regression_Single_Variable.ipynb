```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample data
x_train=np.array([3,6,8,9,12,4,11,7])
y_train=np.array([6,12,16,18,24,8,22,14])
print(f"Shape of x_train and y_train are {x_train.shape} and {y_train.shape}")
```

```python
plt.scatter(x_train,y_train)
plt.xlabel('X')
plt.ylabel('Y')
plt.show
```

```python
# Random w & b
w=1.0
b=0.5
```

```python
def predect(input):
    out=w*input+b
    print(f"predicted output:{out}")
    return out
    
def predict_output(x,w,b):
  m=x.shape[0]
  f_wb=np.zeros(shape=[m])
  for i in range(m):
    f_wb[i]=w*x[i]+b
  return f_wb
```

```python
f_wb=predict_output(x_train,w,b)
plt.scatter(x_train,y_train)
plt.plot(x_train,f_wb)
plt.xlabel('X')
plt.xlabel('Y')
plt.show()
```

```python
def compute_cost(x,y,w,b):
  m=x.shape[0]
  f_wb=predict_output(x,w,b)
  total_cost=0
  for i in range(m):
    total_cost+=(y[i]-f_wb[i])**2
  cost=total_cost/2*m
  return cost

print(f'Cost for w={w} & b={b} is {compute_cost(x_train,y_train,w,b)}')
```

```python
def compute_gradient(x,y,w,b):
    m=x.shape[0]
    g_w=0
    g_b=0
    for i in range (m):
        fw=w*x[i]+b
        fw_i=(fw-y[i])*x[i]
        fb_i=fw-y[i]
        g_w+=fw_i
        g_b+=fb_i
    return g_w/m,g_b/m

def gradient_descent(x,y,iter,alpha):
    global w,b
    iteration=[]
    cost=[]
    for i in range (iter):
        g_w,g_b=compute_gradient(x,y,w,b)
        w=w-alpha*g_w
        b=b-alpha*g_b
        print(f"iteration {i} -> Updated w: {w} b: {b}")
        iteration.append(i)
        cost.append(compute_cost(x,y,w,b))
        return iteration,cost
```

```python
iter,cost=gradient_descent(x_train,y_train,2000,1e-2)
plt.plot(iter,cost)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.show()
```

```python
#Predicting input
val=12
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title('Multiplication by 2')
plt.xlabel('value')
plt.ylabel('result')
x_test=np.array([12,2,5,13,8])
res=predect(val)
f_wb=predict_output(x_test,w,b)
plt.plot(x_test,f_wb)
plt.scatter(x_test,f_wb)
plt.scatter(val,res,color='green')
plt.show()
```
